Summary Statistics 

These are meant to help you understand you dataset in a quick manner.

```python
# Create a copy of the dataframe
summary_df = df.copy()
```

1 Using .describe()

Use .describe() for a quick summary statistics on the dataframe

```python
# Use the .desribe() on the dataframe
summary_df.describe()
```

```python
# Use the function on the column study_hours
summary_df["exam_score"].describe()
```

2 Mean 

The mean is the average and to find that you need to use the function .mean()

```python
# Find the average assignments completed
summary_df['assignments_completed'].mean()
```

3 Median

The median is the middle values when the data is sorted, to get the median you use the .median() function

```python
# Find the median for the exam scores
summary_df['exam_score'].median()
```

4 Mode

The mode is the most common values, to get this value you use the function of .mode()

```python
# Find the mode in the column study_hours
summary_df['study_hours'].mode()
```

5 Minimum and Maximum

The minimum and the maximum are going to be the smallest value or the largest value. To find these you use the .max() and the .min() function

```python
# Find the maximum assignments completed
most_completed = summary_df['assignments_completed'].max()
# Find the minimum assignments completed
least_completed = summary_df['assignments_completed'].min()

most_completed, least_completed
```

6 Range

The range is the difference between the most and least in a column, so to find this you need to subtract the max and the min

```python
# Find the range of the assignments_completed
assignments_range = most_completed - least_completed
assignments_range
```

7 Standard Deviation

The standard deviation tells you how spread apart they are from the mean of the data, this is found by using the .std() function.

```python
# Find the standard deviation for the exam_score
summary_df['exam_score'].std()
```

8 Variance

Variance measures the spread of the column like std(), to find this you need to use the function of .var()

```python
# Find the variance for the exam_scores
summary_df['exam_score'].var()
```

9 Count

Count is used to find the amount of values that are in a column or a dataframe, to do this you need to use the .count()

```python
# Find the amount of values in the exam_score
summary_df['exam_score'].count()
```

```python
# Use the count on the whole dataframe
summary_df.count()
```

10 Value Counts

Values counts how many of that value is showing up in either the dataframe or the colunmn, this is done by using the .value_counts() function

```python
# Find the amt of values are showing up in the dataframe
summary_df['major'].value_counts()
```

11 Grouped Summary Statistics

Grouped summary stats is grouping by a varible and perfroming aggregation on a column based on that grouping. This is done by using the .groupby() function

```python
# Groupby major, find the average exam_score
summary_df.groupby('major')['exam_score'].mean()
```

12 Summary Statistics for Multiple Columns

Using the methods that were previously learned you can use the .agg() function and passing a list allows to do aggregation on many different columns

```python
# Do many aggregations on different columns
summary_df[["study_hours", "quiz_score", "exam_score"]].agg(
    ["count", "mean", "median", "min", "max", "std"]
)
```

13 Correlation as a Summary Statistic

Correlation reveals how two numeric columns move together or how they correlate with each other. This is done by using funciton of .corr()

```python
# How do study_hours, quiz_score, exam_score, 
# and attendance rate correlate
summary_df[["study_hours", "quiz_score", "exam_score", "attendance_rate"]].corr()
```
