```python
# Imports that I will be using

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create the dataframe
df = pd.DataFrame({
    "product_id": [101, 102, 103, 104, 105, 106, 107, 108,
                   109, 110, 111, 112, 113, 114, 115],

    "product_category": ["Electronics", "Clothing", "Home", "Electronics", "Beauty",
                         "Clothing", "Home", "Electronics", "Beauty", "Clothing",
                         "Home", "Electronics", "Beauty", "Clothing", "Home"],

    "region": ["West", "East", "South", "North", "West",
               "South", "East", "West", "North", "East",
               "South", "North", "West", "South", "East"],

    "monthly_sales": [250, 180, 150, 300, 120,
                      210, 160, 330, 140, 220,
                      175, 310, 135, 205, 165],

    "revenue": [12500, 7200, 6000, 15000, 4800,
                8400, 6400, 16500, 5600, 8800,
                7000, 15500, 5400, 8200, 6600],

    "profit": [3200, 1800, 1500, 4100, 900,
               2100, 1600, 4500, 1100, 2300,
               1700, 4300, 1000, 2000, 1550],

    "customer_rating": [4.6, 4.1, 3.9, 4.8, 4.0,
                        4.2, 3.8, 4.9, 4.1, 4.3,
                        3.7, 4.7, 4.2, 4.0, 3.9],

    "returns": [8, 12, 10, 7, 15,
                11, 13, 6, 14, 10,
                12, 7, 16, 11, 13],

    "ad_spend": [2000, 1200, 900, 2500, 800,
                 1300, 950, 2700, 850, 1400,
                 1000, 2600, 900, 1250, 975]
})

df
```

Visual Exploration

Using graphs to quickly try to understand the patterns in your data.

```python
# Create a copy of the dataframe
vis_df = df.copy()
```

1 Bar Chart

Using the bar chart you are able to see count of values and compare with the count of others. This is done by using matplotlib. Specifically the grouped.plot(kind = 'bar')

```python
# Plot product category in a bar chart

# Find the total revenue for the product_category
category_rev = vis_df.groupby('product_category')['revenue'].sum()
# Plot the bar chart
plt.figure()
category_rev.plot(kind = 'bar')
plt.title("Product Category Count")
plt.xlabel("Product_Category")
plt.ylabel("Count")
plt.show()
```

2 Horizontal Bar Chart

The horizontal bar chart accomplishes the same thing as the vertical one, but its useful when the names are much longer. Specifically the grouped.plot(kind = 'barh')

```python
# Find the total profit for each region
profit_region = vis_df.groupby('region')['profit'].sum()
# Plot the horizontal bar chart
plt.figure()
profit_region.plot(kind = "barh")
plt.title("Total Profit by Region")
plt.xlabel("Total Profit")
plt.ylabel("Region")
plt.show()
```

3 Histogram 

The histogram is able to show the distribution of one numeric column. This is done by using matplotlib, but specifilcally the plt.hist(col, bin=#)

```python
# Plot the histogram for the revenue
plt.figure()
plt.hist(vis_df['revenue'], bins = 10)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

4 Boxplot

The boxplot helps you find the spread and the possible outliers. This is done using the matplotlib and more specifically the plt.boxplot(col)

```python
# Plot the boxplot for the profit
plt.figure()
plt.boxplot(vis_df["profit"])
plt.title("Profit Boxplot")
plt.ylabel("Profit")
plt.show()
```

5 Scatter Plot

The scatter plot shows the relationship between two of the numeric columns. This is done using the plt.scatter(col1, col2)

```python
# Plot the scatter plot for the ad spent and the revenue
plt.figure()
plt.scatter(vis_df["ad_spend"], vis_df["revenue"])
plt.title("Ad Spend vs Revenue")
plt.xlabel("Ad Spend")
plt.ylabel("Revenue")
plt.show()
```

6 Line Chart

The line trend is the best for the trends over time. This is done by using the plt.plot(col1, col2, marker = "")

```python
# Plot the product_df over the monthly sales
plt.figure()
plt.plot(vis_df["product_id"], vis_df["monthly_sales"], marker="o")
plt.title("Monthly Sales by Product ID")
plt.xlabel("Product ID")
plt.ylabel("Monthly Sales")
plt.show()
```

7 Pie Chart

The pie chart is able to show the part of a whole. This is done by df.plot(kind = "pie, autopct="%1.1f%%"), and finding a grouping.

```python
# Find the total revenue for each product_category
category_revenue = vis_df.groupby("product_category")["revenue"].sum()
# Plot the piechart
plt.figure()
category_revenue.plot(kind="pie", autopct="%1.1f%%")
plt.title("Revenue Share by Product Category")
plt.ylabel("")
plt.show()
```

8 Grouped Bar Chart

The grouped bar chart compares categories across another category. This could be done by creating a pivot table, and then plotting.

```python
# Create a pivot that has the 
# revenue by the product and region

category_region_revenue = vis_df.pivot_table(
    values = 'revenue',
    index = 'product_category',
    columns = 'region',
    aggfunc = "sum"
)

category_region_revenue.plot(kind = "bar")

plt.title("Revenue by Category and Region")
plt.xlabel("Product Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

9 Correlation Heatmap

A heat map is able to help you see the relationships that are between the numeric columns, this is done by creating a correlation and saving and then plotting it using .imshow()

```python
# Plot the heat map using all of the numeric colomns
numeric_cols = ["monthly_sales", "revenue", "profit", "customer_rating", "returns", "ad_spend"]

correlation = vis_df[numeric_cols].corr()

plt.figure()
plt.imshow(correlation)
plt.colorbar()
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=45)
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.title("Correlation Heatmap")
plt.show()
```

10 Mulitple Histograms

Using the multiple histograms you can quickly check the distributions for multiple numeric columns. This is done by using a for loop to create mulitple different graphs

```python
# Use a loop to create many 
# different histograms to see the distributions

numeric_cols = ["monthly_sales", "revenue", "profit", "customer_rating", "returns", "ad_spend"]

for col in numeric_cols:
    plt.figure()
    plt.hist(vis_df[col], bins=6)
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

11 Multiple Boxplots

The multiple boxplots you are able to check the spread that is across mulitple numeric columns. This is done by ploting and passing through a list of columns.

```python
# Create a muliple different boxplot with different columns
plt.figure()
plt.boxplot([
    vis_df['monthly_sales'],
    vis_df['returns'],
    vis_df['customer_rating']
])
plt.xticks([1, 2, 3], ['monthly_sales', 'returns', 'customer_rating'])
plt.title("Spread Comparison")
plt.ylabel("Values")
plt.show()
```

12 Visual Exploration with Pandas Built-In Plots

pandas has its own built in plotting methods that uses the dataframe itself.

```python
# Histogram 
vis_df["revenue"].plot(kind="hist", bins=6, title="Revenue Distribution")
```

```python
# Bar Chart
vis_df.groupby("product_category")["profit"].sum().plot(
    kind="bar",
    title="Profit by Category"
)
```

```python
# Scatter plot
vis_df.plot(
    kind="scatter",
    x="ad_spend",
    y="revenue",
    title="Ad Spend vs Revenue"
)
```
