```python
# Imports that we will be using
import pandas as pd
import numpy as np
```

```python
# Create the data
df = pd.DataFrame({
    "student_id": [101, 102, 103, 104, 105, 106, 107, 108],
    "name": ["Alex", "Maria", "James", "Sofia", "Daniel", "Emma", "Chris", "Olivia"],
    "age": [20, 21, 19, 22, np.nan, 20, 23, 21],
    "major": ["Business", "Data Science", "Computer Science", "Business", "Math", "Data Science", "Math", None],
    "gpa": [3.4, 3.8, 3.1, 3.6, 2.9, np.nan, 3.7, 3.5],
    "credits_completed": [45, 60, 30, 75, 90, 50, 80, 65],
    "graduated": [False, False, False, False, True, False, True, False]
})

df
```

Dataset overview methods in Pandas

```python
# Create a copy of the dataframe
over_df = df.copy()
```

1 Viewing the first rows

To view the first 5 rows you can use the function .head(), by passing a number as the first argument you can change the amount of row shown

```python
# Show the first 5 rows
over_df.head()
```

```python
# Show the first 10 rows
over_df.head(10)
```

2 Viewing the last rows

To view the last rows in the dataframe you need to use the .tail() fucntion, which you can pass a number as the first argument to change the amount of rows

```python
# View the last 5 rows in the dataframe
over_df.tail()
```

```python
# Viewing the last 3 rows in the dataframe
over_df.tail(3)
```

3 Check the shape of the DataFrame

Checking the shape of the Dataframe you need to use the function of .shape

```python
# Check the shape of the dataframe
over_df.shape
# (8, 7) 8 is the rows and then 7 is columns
```

4 Viewing the Column Names

To view the columns you need to use the .columns function

```python
# Check the columns in the dataframe
over_df.columns
```

```python
# Put the columns into a list for further checking
over_df_columns = over_df.columns.tolist()
# Check the list
over_df_columns
```

5 Checking Data types

Checking data types in the dataframe you need to use the function of .dtypes

```python
# Check the datatypes 
over_df.dtypes
```

6 Get General dataset information

To get the general dataset information you need to use the function .info()

```python
# Check the general information of the dataframe
over_df.info()
```

7 Summary Stats for numeric columns

For summary stats for the numeric columns you can use the function .describe()

```python
# Check the summary stats for the dataframe
over_df.describe()
```

8 Summary Stats for All Columns

To find the summary for All of the columns use the function of .describe() but include the argument of include="all"

```python
# Check the summary stats for all of the dataframe
over_df.describe(include = "all")
```

9 Check Missing Values

To check the missing values, you use the fucntion of .isna(), and the function of .sum()

```python
# To check the for the na rows
over_df.isna().sum()
```

```python
# Another way to check for the NA rows
over_df.isnull().sum()
```

10 Check Percentage of Missing Values

To find the percentage of missing values you use the function of .isna() and then the .mean() then times it by 100

```python
# Find the percentage of missing values in the dataframe
over_df.isna().mean() * 100
```

11 Check the duplicate rows

To check the duplicate rows you can use the .duplicate() and .sum() functions to find the count of the exact duplicate rows

```python
# Check for duplicate rows
over_df.duplicated().sum()
```

12 View Unique Values in a Column

View the unique values that are found in a specific column, you use filtering and then the function of .unique()

```python
# Find the unique values in the column major
over_df['major'].unique()
```

13 Count of Unique values in each columnn

To count the number of unique values in each of the columns you need to use the function of .nunique()

```python
# Find the number of unique vales in each of the cols
over_df.nunique()
```

14 Count values in a column

Counting the values in a column you do filtering and add the function of .value_counts()

```python
# Count the values that are in the column of major
over_df['major'].value_counts()
```

```python
# Adding the dropna=False argument includes the missing values
over_df['major'].value_counts(dropna = False)
```

15 Check index information

Checking the index information you need to use the function of .index

```python
# Check the index info
over_df.index
```

```python
# To reset the index after messy filtering or sorting
over_df.reset_index(drop=True)
```
