Correlational Analysis

This is an analysis to see whether two variables, two columns correlation with each other. The correlation will typlicall range from -1 to 1 which the -1 being a negative correlation and then 1 being a positive correlation

```python
# Create a copy of the dataframe
corr_df = df.copy()
```

1 Correlation Martix

A correlation matrix essentially shows how every numeric column will relate to the other numeric column, and this is done by using function of .corr() and passing the argument of numeric_only=True

```python
# Create a correlation matrix
matrix = corr_df.corr(numeric_only = True)
matrix
```

2 Correlation with only One Target Column

Using the .corr() function you are able to focus on a specific column and see how that column correlates with others

```python
# Create Correlation Matrix focused on sleep_hours
corr_df.corr(numeric_only = True)['sleep_hours']
```

```python
# Can also sort by .sort_values() to have the highest correlation be at the top
corr_df.corr(numeric_only = True)['sleep_hours'].sort_values(ascending=False) 
```

3 Correlation Between Two Columns

Checking the correlation between two columns using the .corr() function

```python
# Check the correlation between attendnce_rate and then final_score
corr_df['attendance_rate'].corr(corr_df['final_score'])
```

4 Pearson Correlation

The pearson correlation is the default method in Pandas, this checks for a linear relationship. This is done by using the .corr() function but you pass the arguments of method="pearson", and add the numeric_only = True

```python
# Create correlation matrix using the pearson method
corr_df.corr(method = "pearson", numeric_only = True)
```

5 Spearman Correlation

This correlation checks whether the variable move in the same direction based on the rank. Use this when the relationship is not perfectly linear but still goes up and down. This is done by using the .corr() function and passing the arguemnt method = "spearman"

```python
# Create a correlation matrix using the spearman method
corr_df.corr(method = "spearman", numeric_only = True)
```

6 Kendall Correlation

The Kendall Correlation also measures the ranked relationships, and is used mainly whent he dataset is small, or when you care about the ordered relationship. This is done by using the .corr() function, and by passing the argument method = "kendall" and the numeric_only = True

```python
# Create a correlation matrix using the kendall method
corr_df.corr(method = "kendall", numeric_only = True)
```

7 Heatmap Visual

Heatmap plots help you see which variable have a strong stronger or weaker relationship, this is created using matplotlib and a correlation matrix.

```python
# Create the correlational matrix and then create the heatmap

# Create the correlational matrix
corr_matrix = corr_df.corr(numeric_only = True)
# Create the heatmap from the corr_matrix
plt.figure()
plt.imshow(corr_matrix)
plt.colorbar()
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation = 45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title("Correlational Heatmap")
plt.show()
```

8 Scatter Plot

The scatter plot helps you visually inspect the relationship between the two variables. Using the matplotlib.

```python
# Create a scatter plot, with the variables of study_hours and then final_score

plt.figure()
plt.scatter(corr_df['study_hours'], corr_df['final_score'])
plt.xlabel("Study Hours")
plt.ylabel("Final Score")
plt.title("Study Hours vs Final Score")
plt.show()
```

9 Negative Correlation Examples

```python
# Plot the scatter plot of a negative correlation
plt.figure()
plt.scatter(df["screen_time_hours"], df["final_score"])
plt.xlabel("Screen Time Hours")
plt.ylabel("Final Score")
plt.title("Screen Time vs Final Score")
plt.show()
```

10 Finding Strong Correlations

Finding strong correlation you need to be able to filter for strong connections. This is done by filtering for the stronger connections that have a correlatio higher or lower than a certain threshold

```python
# Find the stronger correlation with the threshold of 0.7
# Create the correlational matrix
correlation_matrix = df.corr(numeric_only=True)
# Filter for strong correlations
strong_correlations = correlation_matrix["final_score"][
    correlation_matrix["final_score"].abs() > 0.7
]
# Show the stronger correlations
strong_correlations
```

11 Drop ID columns before Correlation

Recommended to Drop ID columns since they should no be included in the correlation anlysis. This is done by usign the .drop(), function.

```python
# Drop the student_id, column and then do the correlation analysis
# Dropping the student_id and save it to a df
no_id_df = corr_df.drop(columns = ['student_id'])
# Create a correlational matrix on the dataframe
no_id_df.corr()
```
