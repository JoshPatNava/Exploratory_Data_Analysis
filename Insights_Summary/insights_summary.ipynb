Insights Summary

Insights Summaries refer to using the the raw data to create a summary that will answer a question.

```python
# Create a copy of the dataframe
insights_df = df.copy()
# Check the columsn
insights_df.columns
```

1 Overall KPI Summary

The overall KPI summary or Key Performance Indicators refer to basic indicators that indicate how the buiness is performing.

```python
# Find the total unique members
total_mem = insights_df['member_id'].nunique()
# Find the total momthly revenue
total_monthly_revenue = insights_df['monthly_fee'].sum()
# Find the average monthly fee
average_monthly_fee = insights_df['monthly_fee'].mean() 
# Find the average visits
average_visits = insights_df['visits_per_month'].mean()
# Find the average satifactions
average_satisfaction = insights_df["satisfaction_score"].mean()
# Find the rate that members are cancelling
cancellation_rate = (df["cancelled"] == "Yes").mean() * 100

# Create a dataframe with all of this information
summary_kpis = pd.DataFrame({
    "metric": [
        "Total Members",
        "Total Monthly Revenue",
        "Average Monthly Fee",
        "Average Visits Per Month",
        "Average Satisfaction Score",
        "Cancellation Rate"
    ],
    "value": [
        total_mem,
        total_monthly_revenue,
        average_monthly_fee,
        average_visits,
        average_satisfaction,
        cancellation_rate
    ]
})

summary_kpis
```

2 Membership Type Summary

The membership type summary will help you compare the Basic, Premium, and the VIP members.

```python
# Create a member ship summary
membership_summary = (
    insights_df.groupby('membership_type')
    .agg(
        member_count=("member_id", "count"),
        total_monthly_revenue=("monthly_fee", "sum"),
        average_visits=("visits_per_month", "mean"),
        average_classes=("classes_attended", "mean"),
        average_satisfaction=("satisfaction_score", "mean")                  
        )
        .reset_index()
        .sort_values(by = "total_monthly_revenue", ascending = False)
)

membership_summary
```

3 Signup Channel Summary

This will show which signup source bring the in the most memebers

```python
# Create a signup channel summary
channel_summary = (
    insights_df.groupby("signup_channel")
      .agg(
          member_count=("member_id", "count"),
          total_monthly_revenue=("monthly_fee", "sum"),
          average_satisfaction=("satisfaction_score", "mean"),
          average_visits=("visits_per_month", "mean")
      )
      .reset_index()
      .sort_values("member_count", ascending=False)
)

channel_summary
```

4 Age group Summary

The Age Group Summary will help you understand which of the age groups are going be the most active.

```python
# Create the Age Group Summary

age_summary = (
    df.groupby("age_group")
      .agg(
          member_count=("member_id", "count"),
          average_visits=("visits_per_month", "mean"),
          average_classes=("classes_attended", "mean"),
          average_satisfaction=("satisfaction_score", "mean")
      )
      .reset_index()
      .sort_values("average_visits", ascending=False)
)

age_summary
```

5 Cancellation Rate Summary

This will help you find which membership type may have the retention problems.

```python
# Create the cancellation rate summary
cancel_rate_by_membership = (
    df.groupby("membership_type")["cancelled"]
      .apply(lambda x: (x == "Yes").mean() * 100)
      .reset_index(name="cancellation_rate_percent")
      .sort_values("cancellation_rate_percent", ascending=False)
)

cancel_rate_by_membership
```

6 Cacellation Rate by Membership Type

Helps to see whether the certain signup channel bring in member who cancel more often.

```python
# Create a cancellation rate by member ship type
cancel_rate_by_channel = (
    df.groupby("signup_channel")["cancelled"]
      .apply(lambda x: (x == "Yes").mean() * 100)
      .reset_index(name="cancellation_rate_percent")
      .sort_values("cancellation_rate_percent", ascending=False)
)

cancel_rate_by_channel
```

7 Find Top Membership Type by Revenue

```python
# Find the top membership type by the revenue
top_membership = membership_summary.loc[
    membership_summary["total_monthly_revenue"].idxmax()
]

top_membership
```

8 Find Most Active Age Group

```python
# Find the most active age group
most_active_age_group = age_summary.loc[
    age_summary["average_visits"].idxmax()
]

most_active_age_group
```

9 Add percent of Total Revenue

```python
membership_summary["percent_of_total_revenue"] = (
    membership_summary["total_monthly_revenue"] /
    membership_summary["total_monthly_revenue"].sum()
) * 100

membership_summary
```

10 Create a Final Insights Table

```python
insights = pd.DataFrame({
    "insight_type": [
        "Top Revenue Membership",
        "Most Active Age Group",
        "Top Signup Channel",
        "Highest Cancellation Membership",
        "Highest Satisfaction Membership"
    ],
    "insight": [
        f"{membership_summary.iloc[0]['membership_type']} generated the most monthly revenue.",
        f"{age_summary.iloc[0]['age_group']} had the highest average visits per month.",
        f"{channel_summary.iloc[0]['signup_channel']} brought in the most members.",
        f"{cancel_rate_by_membership.iloc[0]['membership_type']} had the highest cancellation rate.",
        f"{membership_summary.sort_values('average_satisfaction', ascending=False).iloc[0]['membership_type']} had the highest satisfaction score."
    ]
})

insights
```

11 Visualize Monthly Revenue by Membership Type

```python
plt.figure()
plt.bar(
    membership_summary["membership_type"],
    membership_summary["total_monthly_revenue"]
)
plt.xlabel("Membership Type")
plt.ylabel("Total Monthly Revenue")
plt.title("Monthly Revenue by Membership Type")
plt.show()
```

12 visualize Average Visits by Age Group

```python
plt.figure()
plt.bar(
    age_summary["age_group"],
    age_summary["average_visits"]
)
plt.xlabel("Age Group")
plt.ylabel("Average Visits Per Month")
plt.title("Average Visits by Age Group")
plt.show()
```
