Grouping Aggregation

Grouping means splitting your datastet into groups based upon the columns chosen, and having a summary for a specific aggregation

```python
# Create a copy of the dataframe
group_df = df.copy()
```

1 Group by One Column

To group by one column you use the .groupby() function and pass a column, and mulitply by another column and add an aggregation

```python
# Group by the region, find the total revenue
group_df.groupby('region')['revenue'].sum()
```

2 Group by One Column and Calculate Average

To group by one column and to caluclate the average you use the .groupby() function and use the aggregation of .mean()

```python
# Group by the sales_channel 
# and find the average revenue
group_df.groupby("sales_channel")['revenue'].mean()
```

3 Group by Mulitple Columns

Grouping by mulitple columns you use the .groupby() but pass a list of columns, and then multiply by a column and add a aggregation to it

Note: Order does matter

```python
# Group by both the region and product_catoegory 
# find the total revenue
group_df.groupby(['region', 'product_category'])['revenue'].sum()
```

4 Count Rows in Each group

Counting the rows in each of the columns you use the function .groupby() and .size()

```python
# Find the rows for the group region
group_df.groupby('region').size()
```

5 Count Non-Missing Values

Counting the non-missing values you you use the .groupby() and multiply the by a column and add the aggregation of .count()

```python
# Group by the region count the rows in the revenue
group_df.groupby("region")["revenue"].count()
```

6 Count Unique Values

Counting the unique values you need to use .groupby(), multi column, and then .nunique()

```python
# Groupby the region and find 
# the number of unique customers
group_df.groupby("region")["revenue"].count()
```

7 Use Multiple Aggregation on One column

Using multiple aggregation on one column you need to use the .groupby(), and then multi by a column, and then the .agg() function but pass a list of aggregations

```python
# group by the product category, a 
# find the total, avg, etc revenue
group_df.groupby("product_category")["revenue"].agg(["sum", "mean", "min", "max"])
```

8 Use different aggregations for Different Columns

Using different aggregations for different columns, you do the .groupby() and then use .agg() but pass a dictionary of "column" : "aggregation"

```python
group_df.groupby("region").agg({
    "revenue": "sum",
    "units_sold": "sum",
    "unit_price": "mean",
    "customer_id": "nunique"
})
```

9 Rename Aggregated Columns

Renaming the aggregated columns you do the .groupby(), then the .agg() but pass these name = ('column', 'agg')

```python
# Create the summary table
summary = group_df.groupby("region").agg(
    total_revenue=("revenue", "sum"),
    average_revenue=("revenue", "mean"),
    total_units=("units_sold", "sum"),
    unique_customers=("customer_id", "nunique"),
    order_count=("order_id", "count")
).reset_index()

summary
```

10 Filter Group With GroupBy

After grouping and saving it, you can use filtering on the saved variable

```python
# Group by the product_category 
# and then find the total revenue
category_revenue = group_df.groupby("product_category")["revenue"].sum()
# Filter for those that have a 
# higher revenue than 2000
category_revenue[category_revenue > 2000]
```

11 Sort Grouped Results

To sort by the grouped results, you use the typical structure, and then you add the .sort_values()

```python
# Group by region, find the 
# total revenue and sort in decending order
group_df.groupby("region")["revenue"].sum().sort_values(ascending=False)
```

12 Use as_index = False

Using the argument as_index = False you keep the grouped column as regular column

```python
# Group by region and find the tolal 
# revenue keep grouped column a column 
group_df.groupby("region", as_index=False)["revenue"].sum()
```

13 Transform Grouped Values Back to Original rows

Transforming grouped values to the original rows you do the same structure but you add the .transform() funcntion and pass the aggregation

```python
# Group Region, and find revenue total compare to original
group_df["region_avg_revenue"] = group_df.groupby("region")["revenue"].transform("mean")

group_df[["order_id", "region", "revenue", "region_avg_revenue"]]
```

14 Rank within Groups

TO have ranks within the groups you can use the same structure as before but add the .rank() function

```python
group_df["revenue_rank_in_region"] = group_df.groupby("region")["revenue"].rank(
    ascending=False
)

group_df[["order_id", "region", "revenue", "revenue_rank_in_region"]]
```

15 Get the Largest or Lowest value in each group 

To get the lowest or the largest you can ue the function .idmax() or .idxmin() at the end.

```python
# For the largest
group_df.loc[group_df.groupby("region")["revenue"].idxmax()]
```

```python
# For the lowest
group_df.loc[group_df.groupby("region")["revenue"].idxmax()]
```

17 Pivot Table as Grouped Summary 

Pivot table is another way to have a grouped summary

```python
# Create a pivot table
pd.pivot_table(
    group_df,
    values="revenue",
    index="region",
    columns="product_category",
    aggfunc="sum"
)
```
